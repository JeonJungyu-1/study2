```python
def decimal(a):
    for i in range(2, a):
        if a % i == 0:
            return False
    return True
if decimal(2):
    print('소수입니다')
else:
    print('소수가 아닙니다')
```

```python
def decimal2(a):
    cnt = 1
    for i in range(3, a+1):
        chk = False
        for j in range(2, i):
            if i % j == 0:
                chk = True
                break
        if chk == False:
            cnt += 1
    return cnt

print(decimal2(13))
```

```python
def decimal2(a):
    cnt = 0
    for i in range(2, a+1):
        if not decimal(i):
            continue
        cnt += 1
    return cnt
print(decimal2(2))
```
